### Quanto bene risolve il nostro problema il metodo di Eulero?

Provate a modificare il valore di $k$ e a rieseguire le celle del notebook visto.
Se scelgliete $k=500$ potrete notare che la soluzione numerica è molto distante da quella analitica in questo caso.

Provate a diminuire il passo: $h = 0.001$. 
Vedrete che la soluzione numerica è sufficientemente accurata a costo di diminuire molto il passo di integrazione.

Il problema test in questo caso è detto *Stiff*, ne analizzeremo le caratteristiche più nel dettaglio in seguito.

```python
## import delle Librerie
import numpy as np
import matplotlib.pyplot as plt
```

Soluzione errata dovuta ad un passo troppo grande:

```python
# Problema TEST e Soluzione analitica:
k = 500.0
func_test = lambda y: -k *y;
sol_real = lambda y0,t0,t: y0*np.exp(-k*(t-t0))

# Metodo di Eulero Esplicito o "in avanti" (EE)
h  = 0.01; # passo
tf = 0.1
times = np.arange(0,tf,h)
u_EE  = np.zeros(len(times)); # array che conterra' la soluzione
u_n   = 1.0
u_EE[0] = u_n
for i in range(0,len(times)-1):
    u_n1      = u_n + h*func_test(u_n);
    u_EE[i+1] = u_n1
    u_n       = u_n1
#end
```

```python
# GRAFICO
y_real = sol_real(u_EE[0],times[0],times)
plt.figure(figsize=(7,5))
plt.plot(times, u_EE,'o-',label='y_EE: sol numerica (EE)')
plt.plot(times, y_real ,'g--',linewidth=3,label='sol reale con y(%s)=y_EE[%s]'%(times[0],0))

plt.legend(loc='center left', bbox_to_anchor=(1, 0.5));
plt.title("Soluzione reale e numerica col metodo di Eulero Esplicito \n della ODE:  y'=%sy,  y(0)=1 e h=%s" % (int(k),h),fontweight="bold")
```
